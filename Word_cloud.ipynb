```python
import urllib.request

response = urllib.request.urlopen('https://apple.com/')
html = response.read()
print(len(html))
```

```python
html_content = html.decode('utf-8')
token = [tok for tok in html_content.split()]
#html_content.split()
print(token)
print(len(token))
```

```python
import re
html_content = html.decode('utf-8')
tokens = re.split('\W+',html_content)
print(tokens)
```

```python
from bs4 import BeautifulSoup
import nltk

html_content = html.decode('utf-8')

soup = BeautifulSoup(html_content)
clean_text = soup.get_text()

tokens = [tok for tok in clean_text.split()]
print(tokens)
```

```python
freq_dist = {}
for tok in tokens:
  if tok in freq_dist:
    freq_dist[tok] = freq_dist[tok] +1
  else:
    freq_dist[tok] = 1
  print(freq_dist)
```

```python
freq_dist_nltk = nltk.FreqDist(tokens)

for word, freq in freq_dist_nltk.items():
  print(f"{word}:{freq}")
```

```python
freq_dist_nltk.plot(50, cumulative=False)
```

```python
from nltk.corpus.reader.wordnet import FreqDist
from nltk.corpus import stopwords
stopword_list = set(stopwords.words('english'))
clean_tokens = [tok for tok in tokens if len(tok.lower())>1 and (tok.lower() not in stopword_list)]
print(len(clean_tokens))
print(clean_tokens)
freq_dist_nltk = FreqDist(clean_tokens)
freq_dist_nltk.plot(50, cumulative=False)
```

```python
import wordcloud
from wordcloud.wordcloud import WordCloud
from wordcloud import WordCloud
import matplotlib.pyplot as plt

word_freq_dict = {word: freq for word, freq in freq_dist_nltk.items()}
word_cloud = WordCloud(width= 800,
                       height=800,
                       background_color='white',
                       min_font_size=10)
word_cloud.generate_from_frequencies(word_freq_dict)
plt.Figure()
plt.imshow(word_cloud)
plt.axis('off')
plt.tight_layout(pad=0)
plt.show()
```
